```python
import pandas as pd
df = pd.read_csv("C:/Users/Admin/Documents/Datasets/Cars93.csv")
df.drop(labels=["Unnamed: 0"], axis=1, inplace=True)
df.head()
```

```python
df[["Luggage.room"]] = df[["Luggage.room"]].fillna(13.890244)
df[["Rear.seat.room"]] = df[["Rear.seat.room"]].fillna(df[["Rear.seat.room"]].mean())
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
cat = []
con = []
for i in df.columns:
    if df[i].dtypes=="object":
        cat.append(i)
        sb.countplot(df[i], hue=df.AirBags)
    else:
        con.append(i)
        sb.boxplot(df.AirBags, df[i])
    plt.show()
```

```python
# for SGD  =>  df_1 = df[["Type", "EngineSize", "Horsepower", "Rev.per.mile", "Passengers", "Length", "Wheelbase", "Luggage.room", "Origin"]]
df_1 = df[["Type", "Min.Price", "MPG.city", "DriveTrain", "Cylinders", "EngineSize", "Horsepower", "Rev.per.mile", "Man.trans.avail", "Passengers", "Length", "Wheelbase", "Width", "Luggage.room", "Origin"]]
```

```python
cat=[]
con=[]
for i in df_1.columns:
    if df_1[i].dtypes == "object":
        cat.append(i)
    else:
        con.append(i)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
B = pd.DataFrame(ss.fit_transform(df_1[con]), columns=con)
B.columns
```

```python
B1 = pd.get_dummies(df_1[cat], drop_first=True)
B1.columns
```

```python
X = B1.join(B)
y = df.AirBags
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)
```

```python
from sklearn.neural_network import MLPClassifier
mlpc = MLPClassifier(random_state=20, solver="adam", hidden_layer_sizes=(10, 10), max_iter=1000)
model = mlpc.fit(X_train, y_train)
pred = model.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred) #0.6842105263157895
```
