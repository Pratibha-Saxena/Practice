```python
import pandas as pd
df = pd.read_csv("C:/Users/Admin/Documents/Datasets/Credit.csv")
df.head()
```

# Divide columns into cat | con

```python
cat = []
con = []
for i in df.columns:
    if df[i].dtypes == "object":
        cat.append(i)
    else:
        con.append(i)
```

# One hot encoding of all cat columns (Except y column)

```python
cat.remove("Student")
```

```python
OHEcols = pd.get_dummies(df[cat])
```

# Join these columns with continuous columns

```python
df[con]
con.remove("Unnamed: 0")
con.remove("ID")
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
df[con] = ss.fit_transform(df[con])
```

```python
X = df[con].join(OHEcols)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = df[["Student"]]
y = y.apply(le.fit_transform)
```

```python
X
```

# Divide data into training and testing split

```python
from sklearn.model_selection import train_test_split
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=60)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)
y_train.value_counts()
```

# Create a Neural Network

```python
from keras import Sequential
from keras.layers import Dense
```

```python
# last time we did like dis
from sklearn.neural_network import MLPClassifier
mlpc = MLPClassifier(hidden_layer_sizes=(100, 120, 50, 30))
```

```python
# Now.. using Keras
model = Sequential()
model.add(Dense(100, activation="relu", input_dim=14))
model.add(Dense(120, activation="relu"))
model.add(Dense(50, activation="relu"))
model.add(Dense(30, activation="relu"))
model.add(Dense(1, activation="sigmoid"))
```

```python
model.compile(optimizer="sgd", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
model_nn = model.fit(X_train, y_train, validation_split=0.1, epochs=10)
```

```python
pred_proba = model.predict(X_test)
pred = [int(i>0.5) for i in pred_proba]
# pred1 = []
# for i in pred:
#    if(i > 0.5):
#        pred1.append(1)
#    else:
#        pred1.append(0)
```
