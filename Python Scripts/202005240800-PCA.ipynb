```python
import pandas as pd
df = pd.read_csv("C:/Users/Admin/Documents/Datasets/Cars93.csv")
df.head()
```

```python
df.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_1 = df.apply(le.fit_transform)
```

```python
model = pca.fit(df_1)
```

```python
principle_components = pd.DataFrame(model.components_)
```

```python
principle_components
```

```python
m = 0
for evr in list(model.explained_variance_ratio_):
    m = m+1
    print("principle_components", m, "%.2f"%(evr))
```
