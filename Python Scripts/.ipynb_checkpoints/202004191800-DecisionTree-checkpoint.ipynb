```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
X = df.drop(labels=["Unnamed: 0", "logwage", "wage", "region"], axis=1)
# wage is target, logwage is derived from wage, region feature has only one value for all obs
X.head()
```

```python
print(X.maritl.unique())
X["maritl"] = le.fit_transform(X["maritl"])
X.head()
```

```python
print(X.race.value_counts())
print(X.education.value_counts())
print(X.region.value_counts())
print(X.jobclass.value_counts())
print(X.health.value_counts())
print(X.health_ins.value_counts())
```

```python
X["race"] = le.fit_transform(X["race"])
X["education"] = le.fit_transform(X["education"])
X["jobclass"] = le.fit_transform(X["jobclass"])
X["health"] = le.fit_transform(X["health"])
X["health_ins"] = le.fit_transform(X["health_ins"])
X.head()
```

```python
y = df.wage
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
model = dtr.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, pred)
```

```python
from sklearn.tree import export_graphviz
export_graphviz(dtr, out_file = '')
```
