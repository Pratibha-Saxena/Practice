```python
# EmployeeCount: 1 everywhere  --> Remove
# Over18: Y everywhere --> Remove
# StandardHours: 80 everywhere --> Remove
# EmployeeNumber: --> Remove

df_dropped = df.drop(labels=["EmployeeCount", "Over18", "StandardHours", "EmployeeNumber"], axis=1)
```

```python
# BusinessTravel: 3-ordinal --> one hot ?  no. making ordinal numeric
# Education: 5-ordinal numeric --> one hot ? no. keeping ordinal
# EnvironmentSatisfaction: 4-ordinal numeric --> one hot ? no. keeping ordinal
# JobInvolvement: 4-ordinal numeric --> one hot ? no. keeping ordinal
# JobLevel: 5-ordinal numeric --> one hot ? no. keeping ordinal
# JobSatisfaction: 4-ordinal numeric --> one hot ? no. keeping ordinal
# RelationshipSatisfaction: 4-ordinal numeric --> one hot ? no. keeping ordinal
# StockOptionLevel: 4-ordinal numeric --> one hot ? no. keeping ordinal
# WorkLifeBalance: 4-ordinal numeric --> one hot ? no. keeping ordinal

df_enc = df_dropped
col_val_order = ["Non-Travel", "Travel_Rarely", "Travel_Frequently"]
df_enc["BusinessTravel"] = df_dropped["BusinessTravel"].apply(lambda x: col_val_order.index(x))
df_enc.head()
```

```python
df_enc.info()
```

```python
cat = []
con = []
for i in df_enc.columns:
    if df_enc[i].dtype == 'object':
        cat.append(i)
    else:
        con.append(i)
print(cat)
print(con)
```

```python
for i in cat:
    cr_tab = pd.crosstab(df_enc[i], df_enc["Attrition"])
    cr_tab['perc'] = cr_tab[1]/(cr_tab[0]+cr_tab[1])
    display(cr_tab)
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
for i in cat:
    sb.countplot(df_enc.loc[df_enc.Attrition==0][i])
    plt.show()
    sb.countplot(df_enc.loc[df_enc.Attrition==1][i])
    plt.show()
```

```python
df_dummies = pd.get_dummies(df_enc,
                         prefix=["Department", "EducationField", "Gender", "JobRole", "MaritalStatus", "OverTime"],
                         columns = ["Department", "EducationField", "Gender", "JobRole", "MaritalStatus", "OverTime"],
                         drop_first=True
                        )
df_dummies.corr().iloc[1:, 0]
```

```python
import pandas as pd
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
X = df_dummies.iloc[:, 1:]  #independent columns
y = df_dummies.iloc[:, 1]    #target column

#apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=43)
fit = bestfeatures.fit(X, y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
print(featureScores.nlargest(43,'Score'))  #print 10 best features
```

```python
X = df_dummies.iloc[:, 1:]  #independent columns
y = df_dummies.iloc[:, 1]    #target column i.e price range
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model = ExtraTreesClassifier()
model.fit(X,y)
feature_scores = pd.DataFrame(model.feature_importances_, columns=["Score"])
feature_scores["Specs"] = X.columns
display(feature_scores.nlargest(43, "Score"))

print(list(zip(X.columns, model.feature_importances_))) #use inbuilt class feature_importances of tree based classifiers
#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=30)
```

```python
X_train.shape
X_test.shape
X_train.columns
```

```python
col_names = [
     "Department",
#     "EducationField", "Gender",
     "JobRole",
     "MaritalStatus", 
     "OverTime"]
X_train = pd.get_dummies(X_train,
                         prefix=col_names,
                         columns = col_names,
                         drop_first=True
                        )
```

```python
X_test = pd.get_dummies(X_test,
                         prefix=col_names,
                         columns = col_names,
                         drop_first=True
                        )
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

tp = {"min_samples_leaf": range(5, 10, 1),
      "random_state": range(0, 30, 5),
      "criterion": ["gini", "entropy"],
     }
# cv = GridSearchCV(DecisionTreeClassifier(),
#                  param_grid=tp, scoring=["precision", "recall", "f1", "roc_auc"], refit="f1", cv=10)
# cv_model = cv.fit(X_train, y_train)

tp = {"min_samples_leaf": range(5, 10, 1),
      "random_state": [25],
      "criterion": ["gini", "entropy"],
     }
# cv = GridSearchCV(RandomForestClassifier(),
#                  param_grid=tp, scoring=["precision", "recall", "f1", "roc_auc"], refit="f1", cv=10)
# cv_model = cv.fit(X_train, y_train)


dtc = DecisionTreeClassifier(min_samples_leaf=9, random_state=25)
dtc_model = dtc.fit(X_train, y_train)

rfc = RandomForestClassifier(n_estimators=200, random_state=25)
rfc_model = rfc.fit(X_train, y_train)

abc = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=1, random_state=25), random_state=25, n_estimators=200)
abc_model = abc.fit(X_train, y_train)

# lr = LogisticRegression(max_iter=4000)
# lr_model = lr.fit(X_train, y_train)

# gnb = GaussianNB()
# gnb_model = gnb.fit(X_train, y_train)
```

```python
pred_dtc = dtc_model.predict(X_test)
pred_rfc = rfc_model.predict(X_test)
pred_abc = abc_model.predict(X_test)
# pred_lr = lr_model.predict(X_test)
# pred_gnb = gnb_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
from sklearn.metrics import classification_report, roc_auc_score, f1_score
print("---DTC---")
# print(accuracy_score(y_test, pred_dtc))
print(precision_score(y_test, pred_dtc))
print(recall_score(y_test, pred_dtc))
print(f1_score(y_test, pred_dtc))
print("---RFC---")
# print(accuracy_score(y_test, pred_rfc))
print(precision_score(y_test, pred_rfc))
print(recall_score(y_test, pred_rfc))
print(f1_score(y_test, pred_rfc))
print("---ABC---")
# print(accuracy_score(y_test, pred_abc))
print(precision_score(y_test, pred_abc))
print(recall_score(y_test, pred_abc))
print(f1_score(y_test, pred_abc))
# print("---LR---")
# # print(accuracy_score(y_test, pred_abc))
# print(precision_score(y_test, pred_lr))
# print(recall_score(y_test, pred_lr))
# print(f1_score(y_test, pred_lr))
# print("---GNB---")
# # print(accuracy_score(y_test, pred_abc))
# print(precision_score(y_test, pred_gnb))
# print(recall_score(y_test, pred_gnb))
# print(f1_score(y_test, pred_gnb))

print(confusion_matrix(y_test, pred_dtc))
print(confusion_matrix(y_test, pred_rfc))
print(confusion_matrix(y_test, pred_abc))
# print(confusion_matrix(y_test, pred_lr))
# print(confusion_matrix(y_test, pred_gnb))
```
