```python
import pandas as pd
df = pd.read_csv("C:/Users/Admin/Documents/Datasets/Restaurant_Reviews.tsv", sep="\t")
df.head()
```

```python
df[0:5][["Review"]].values
```

```python
rev = []
for i in df.Liked:
    if(i==0):
        rev.append("Negative")
    else:
        rev.append("Positive")
df["Liked_str"] = rev
df.head()
```

```python
from preprocessing_NLP import rem_spl_chars,stem,tokenize
new_rev = []
import re
for i in df.Review:
    i = rem_spl_chars(i)
    i = re.sub("[..]", "", i)
    new_rev.append(i)
new_rev
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X = cv.fit_transform(new_rev)
y = df[["Liked_str"]]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
model = nb.fit(X_train, y_train.Liked_str)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print(confusion_matrix(y_test, pred))
print(accuracy_score(y_test, pred))
```

```python
df_sms = pd.read_csv("C:/Users/Admin/Documents/Datasets/spamSMS.csv")
df_sms.head()
```

```python
new_rev = []
for i in df_sms.Message:
    i = rem_spl_chars(i)
    i = re.sub("[..]", "", i)
    new_rev.append(i)
new_rev
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X = cv.fit_transform(new_rev)
y = df_sms.Category
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=90)
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
model = nb.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print(confusion_matrix(y_test, pred))
print(accuracy_score(y_test, pred))
```
