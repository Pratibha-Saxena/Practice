```python
import pandas as pd
df = pd.read_csv("C:/Users/Admin/Documents/Datasets/50_Startups (1).csv")
df.head()
```

```python
df.info()
```

```python
import seaborn as sb
sb.pairplot(df)
```

```python
df.RND.hist()
```

```python
import numpy as np
mean = np.mean(df.PROFIT)
new_profit = []
for i in df.PROFIT:
    if i<mean:
        new_profit.append(0)
    else:
        new_profit.append(1)
df['NP'] = new_profit
```

```python
from sklearn.model_selection import train_test_split
X = df[["MKT", "RND", "ADMIN"]]
y = df[["NP"]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
model = lr.fit(X_train, y_train['NP'])
pred = lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, pred)
```

```python
pred
```

```python
model.predict_proba(X_test)
```

```python
import sklearn.metrics as m
m.accuracy_score(y_test, pred)
```

# calculate the fpr and tpr for all thresholds of the classification

```python
true_probas = model.predict_proba(X_test)[:, 1]
true_probas
```

```python
fpr, tpr, threshold = m.roc_curve(y_test, true_probas)
roc_auc = m.auc(fpr, tpr)
print(fpr)
print(tpr)
print(roc_auc)
```

```python
import matplotlib.pyplot as plt
plt.plot(fpr, tpr, label=roc_auc)
plt.show()
```

```python
roc_auc
```

```python
print(m.classification_report(y_test, pred))
```
