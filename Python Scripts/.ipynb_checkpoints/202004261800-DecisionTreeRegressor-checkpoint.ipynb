```python
import pandas as pd
df = pd.read_csv("C:/Users/Admin/Documents/Datasets/Cars93.csv")
df.head()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
for i in df.columns:
    if df[i].dtypes == 'object':
        sb.countplot(df.Type, hue=df[i])
    else:
        sb.boxplot(df.Type, df[i])
    plt.show()
```

```python
df.columns
```

```python
X = df.drop(labels=["Unnamed: 0", "Type", "Manufacturer", "Model", "Make", "Passengers"], axis =1)
y = df.Type
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X = X.apply(le.fit_transform)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion="entropy", random_state=30)
model = dtc.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

# Trying to compare with the case using LogisticRegressor Classification

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
pred = lr.predict(X_test)
accuracy_score(y_test, pred)
```

# Trying Pruning on the Tree with GridSearchCV

```python
X = df.drop(labels=["Unnamed: 0", "Type", "Manufacturer", "Model", "Make", "Passengers"], axis=1)
y = df.Type
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X = X.apply(le.fit_transform)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)
sb.countplot(y)
plt.show()
sb.countplot(y_train)
plt.show()

from sklearn.model_selection import GridSearchCV
dtc1 = DecisionTreeClassifier()
tp = {"random_state": range(2, 20, 1), "min_samples_leaf": range(2, 20, 1)}
cv = GridSearchCV(dtc1, tp, scoring="accuracy", cv=4)
cv_model = cv.fit(X_train, y_train)
rnd = cv_model.best_params_['random_state']
min_smp_lf = cv_model.best_params_['min_samples_leaf']
print(rnd)
print(min_smp_lf)
cv_model.best_score_
```

```python
dtc1 = DecisionTreeClassifier(random_state=rnd, min_samples_leaf=min_smp_lf)
model1= dtc1.fit(X_train, y_train)
pred1 = model1.predict(X_test)
accuracy_score(y_test, pred1)
```
