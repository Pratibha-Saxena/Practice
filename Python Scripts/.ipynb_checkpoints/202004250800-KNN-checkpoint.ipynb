```python
import pandas as pd
df = pd.read_csv("c:/Users/Admin/Documents/Datasets/Credit.csv")
df.head()
```

# EDA

```python
import seaborn as sb
import matplotlib.pyplot as plt
for i in df.columns:
    if df[i].dtypes == 'object':
        sb.boxplot(df[i], df.Income)
    else:
        sb.scatterplot(df[i], df.Income)
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
# How label encoder standardizes the continuous data automatically....??
# And what effect it has on continuous data....??
le = LabelEncoder()
df_enc = df.apply(le.fit_transform)
```

```python
df_enc.head()
```

```python
sb.boxplot(df.Education, df.Income, whis=5)
plt.show()
# sb.scatterplot(df.)
```

```python
from sklearn.model_selection import GridSearchCV, train_test_split
X = df_enc[["Limit", "Rating", "Cards", "Age", "Education", "Balance"]]
y = df_enc[["Income"]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=25)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
for i in range(2, 20, 1):
    knn = KNeighborsRegressor(n_neighbors=i)
    model = knn.fit(X_train, y_train)
    pred = model.predict(X_test)
    print(i, mean_squared_error(y_test, pred))
```

```python
knn_grd_cv = KNeighborsRegressor()
tp = {"n_neighbors": range(2, 20, 1)}
cv = GridSearchCV(knn_grd_cv, tp, scoring="neg_mean_squared_error", cv=4)
cv_model = cv.fit(X_train, y_train)
print(cv_model.best_params_)
cv_model.best_score_
```

```python
sb.distplot(df.Income)
plt.show()
sb.distplot(newincome)
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix, roc_auc_score, auc
cm = confusion_matrix(y_test, pred)
import numpy as np
import sys
# np.set_printoptions(threshold=1000)
print(list(cm))
print("Classification Report: ", classification_report(y_test, pred))
print("F1Score: ", f1_score(y_test, pred, average="macro"))
print("Precision: ", precision_score(y_test, pred, average="macro"))
print("Recall: ", recall_score(y_test, pred, average="macro"))
print("Accuracy: ", accuracy_score(y_test, pred)*100)
```

# It is Multiclass Classification... bcoz v hav considered income as classes here....

```python
true_probas = model.predict_proba(X_test)[:, 1]
true_probas
fpr, tpr, threshold = roc_auc_score(y_test, true_probas)
auc = auc(fpr, tpr)
print(fpr)
print(tpr)
print(auc)
```
