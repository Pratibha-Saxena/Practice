```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv("C:/Users/Admin/Documents/Datasets/Cars93.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
cat = []
con = []
for i in df.columns:
    if df[i].dtypes == 'object':
        cat.append(i)
    else:
        con.append(i)
```

```python
# Remove high cardinality categorical column before plotting
cat.remove("Manufacturer")
cat.remove("Model")
cat.remove("Make")
```

```python
for i in cat:
    for j in con:
        sb.boxplot(df[i], df[j])
        plt.show()
```

```python
sb.heatmap(df.corr())
```
