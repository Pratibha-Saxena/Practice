```python
import pandas as pd
df = pd.read_csv("C:/Users/Admin/Documents/Datasets/Cars93.csv")
df.head()
```

```python
y = df[["Weight"]]
# X=
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
for i  in df.columns:
    if df[i].dtypes == 'object':
        sb.boxplot(df[i], df.Weight)
    else:
        sb.scatterplot(df[i], df.Weight)
    plt.show()
```

```python
X = df[["Price", "Rev.per.mile", "MPG.city", "EngineSize", "Horsepower", "Fuel.tank.capacity", "Length", "Wheelbase", "Width"]]
y = df["Weight"]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X1 = X.apply(le.fit_transform)
```

```python
X2 = pd.get_dummies(df[["Type", "Cylinders"]])
Xnew = X1.join(X2)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

```python
from sklearn.neural_network import MLPRegressor
mlpr = MLPRegressor(random_state=30, max_iter=3500)
nn = mlpr.fit(X_train, y_train)
pred = nn.predict(X_test)
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, pred)
```

```python
from sklearn.neural_network import MLPRegressor
mlpr = MLPRegressor(solver="sgd", nesterovs_momentum=True, batch_size=20, random_state=30)
nn = mlpr.fit(X_train, y_train)
pred = nn.predict(X_test)
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, pred)
```

```python
from sklearn.neural_network import MLPRegressor
mlpr = MLPRegressor(solver="adam", random_state=30, max_iter=3500)
nn = mlpr.fit(X_train, y_train)
pred = nn.predict(X_test)
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, pred)
# nn.n_iter_
```

```python
from sklearn.neural_network import MLPRegressor
mlpr = MLPRegressor(solver="adam", hidden_layer_sizes=(20), random_state=30, max_iter=17200)
nn = mlpr.fit(X_train, y_train)
pred = nn.predict(X_test)
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, pred)
```

```python
from sklearn.neural_network import MLPRegressor
mlpr = MLPRegressor(solver="adam", hidden_layer_sizes=(20, 20), random_state=30, max_iter=3200)
nn = mlpr.fit(X_train, y_train)
pred = nn.predict(X_test)
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, pred)
```

```python
from sklearn.neural_network import MLPRegressor
mlpr = MLPRegressor(solver="adam", hidden_layer_sizes=(20, 20, 30), random_state=30)
nn = mlpr.fit(X_train, y_train)
pred = nn.predict(X_test)
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, pred)
```

```python
P = list(range(2, 25, 2))
for i in range(2, 10, 2):
    for j in range(2, 10, 2):
        P.append((i, j))
for i in range(2, 10, 2):
    for j in range(2, 10, 2):
        for k in range(2, 10, 2):
            P.append((i, j, k))
```

```python
P
```

```python
tp = {"hidden_layer_sizes": P}
```

```python
from sklearn.model_selection import GridSearchCV
gcv = GridSearchCV(mlpr, tp, scoring="neg_mean_absolute_error", cv=4)
cvmodel = gcv.fit(X_train, y_train)
cvmodel.best_params_
```

```python
from sklearn.neural_network import MLPRegressor
mlpr = MLPRegressor(solver="adam", hidden_layer_sizes=cvmodel.best_params_["hidden_layer_sizes"], random_state=30, max_iter=13300)
nn = mlpr.fit(X_train, y_train)
pred = nn.predict(X_test)
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, pred)
```
