# MULTI-VARIATE OUTLIERS ---> K-MEANS CLUSTERING

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=10, init="random")
df_b = df[["RND", "MKT"]]
model = km.fit(df_b[["RND", "MKT"]])
model.labels_
```

```python
col = []
for i in model.labels_:
    if i == 0:
        col.append("black")
    elif i == 1:
        col.append("blue")
    elif i == 2:
        col.append("purple")
    elif i == 3:
        col.append("red")
    elif i == 4:
        col.append("yellow")
    elif i == 5:
        col.append("green")
    elif i == 6:
        col.append("pink")
    elif i == 7:
        col.append("grey")
    elif i == 8:
        col.append("violet")
    elif i == 9:
        col.append("orange")
```

```python
plt.scatter(df.RND, df.MKT, c=col)
```

```python
df_labels = pd.DataFrame(model.labels_, columns=["KMeansLabels"])
```

```python
df_labels.KMeansLabels.value_counts()
```
