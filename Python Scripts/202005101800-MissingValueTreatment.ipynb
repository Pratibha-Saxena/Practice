# Treating missing data with replacement

```python
import numpy as np
mu = np.mean(df[df["Luggage.room"].notnull()]["Luggage.room"])
```

```python
mu
```

```python
df = df["Luggage.room"].fillna(mu)
```

```python
# Multiple Columns
import numpy as np
def replace_all_na(col):
    mu = df[df[col].notnull()][col]
    print(mu)
    return df[col].fillna(mu)
```

```python
cat = []
con = []
for i in df.columns:
    if df[i].dtypes == 'object':
        cat.append(i)
    else:
        con.append(i)
```

```python
list(map(replace_all_na, con))
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(strategy="mean")
pd.DataFrame(imp.fit_transform(df[con]), columns=con)
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(strategy='most_frequent')
pd.DataFrame(imp.fit_transform(df[cat]), columns=cat)
```
